```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Impact_of_Remote_Work_on_Mental_Health.csv')
df.head()
```

```python
df.info()
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns

for column in categorical_columns:
    print(f"Categories in '{column}':")
    print(df[column].unique(), "\n")
```

```python
df_cleaned = df.dropna()
df_cleaned.info()
```

```python
categorical_columns = df_cleaned.select_dtypes(include=['object']).columns

for column in categorical_columns:
    print(f"Categories in '{column}':")
    print(df_cleaned[column].unique(), "\n")
```

```python
data_filtered = df_cleaned[df_cleaned['Gender'] != 'Prefer not to say']
data_filtered.info()
```

```python
categorical_columns = data_filtered.select_dtypes(include=['object']).columns

for column in categorical_columns:
    print(f"Categories in '{column}':")
    print(data_filtered[column].unique(), "\n")
```

```python
data_filtered.to_csv('new_df.csv')
```
